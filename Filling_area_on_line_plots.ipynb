```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('c:/users/Seera Bhargavi/Desktop/python/matplotlib/data.txt')
data.head()
```

```python
ages=data['Age']
dev_salaries=data['All_Devs']
python_salaries=data['Python']
js_salaries=data['JavaScript']
```

```python
plt.plot(ages,dev_salaries,label='All Devs',color='r',linestyle='--')
plt.plot(ages,python_salaries,label='Python',color='b',linestyle=':')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```

```python
plt.plot(ages,dev_salaries,label='All Devs',color='r',linestyle='--')
plt.plot(ages,python_salaries,label='Python',color='b',linestyle=':')
plt.fill_between(ages,python_salaries,alpha=0.25)
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```

```python
plt.plot(ages,dev_salaries,label='All Devs',color='k',linestyle='--')
plt.plot(ages,python_salaries,label='Python',color='b',linestyle='-')
overall_median=50000
plt.fill_between(ages,python_salaries,overall_median,alpha=0.25)
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.show()
```

#### Condition

```python
plt.plot(ages,dev_salaries,label='All Devs',color='k',linestyle='--')
plt.plot(ages,python_salaries,label='Python',color='b',linestyle='-')
overall_median=50000
plt.fill_between(ages,python_salaries,overall_median,alpha=0.25)
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.fill_between(ages,python_salaries,overall_median,where=(python_salaries > overall_median),alpha=0.25,label='Above average')
plt.show()
```
