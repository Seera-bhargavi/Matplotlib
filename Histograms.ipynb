```python
import matplotlib.pyplot as plt
```

#### creating histogram using hist() function

```python
#defining ages
ages=[18,19,21,25,26,26,30,32,38,45,55]
```

```python
# hist() is used to create histogram and it takes array as argument
plt.hist(ages,bins=5,edgecolor='black')
plt.show()
```

```python
# We can define bins separately as list and pass as an argument
bins=[10,20,30,40,50,60]
# Now it will take 10-20 as interval, 20-30 as interval, and so on,,.
plt.hist(ages,bins=bins,edgecolor='black')
plt.show()
```

```python
# Based on bins only it will plot
# eg: if we give bins from 20 it will not include ages 18,19
bins=[20,30,40,50,60]
plt.hist(ages,bins=bins,edgecolor='black')
plt.show()
```

#### Plot the vertical line

```python
ages=[18,19,21,25,26,26,30,31,32,35,36,38,38,45,55]
bins=[10,20,30,40,50,60]
plt.hist(ages,bins=bins,edgecolor='black')
median_age=29
color='red'
plt.axvline(median_age,color=color,linewidth=2)
```
