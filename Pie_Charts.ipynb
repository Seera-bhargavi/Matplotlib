```python
#importing libraries
import matplotlib.pyplot as plt
import numpy as np
```

#### Let's see simple pie-chart

```python
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# one piece we call wodge
# By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:
```

#### Labels

```python
# Add labels to the pie chart with the label parameter
# The label parameter must be an array with one label for each wedge

y=np.array([35,25,25,15])
lables=['A','B','C','D']
plt.pie(y,labels=lables)
plt.show()
```

#### startangle

```python
# Default start angle is at x-axis,but you can change the startangle by adding parameter startangle with value
# i.e you want to start your pie-chart at that angle

plt.pie(y,labels=lables,startangle=45)
```

#### wedgeProps

```python
plt.pie(y,labels=lables,startangle=45,wedgeprops={'edgecolor':'black'})
```

#### Adding colors to wedges

```python
#colors=['blue','orange','yellow','green']
# You can also use hexa values
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']
plt.pie(y,labels=lables,colors=colors,startangle=45)
plt.show()
```

#### Reading data from csv file and representing in a pie-chart

```python
import pandas as pd
from collections import Counter
```

```python
df=pd.read_csv('c:/users/Seera Bhargavi/Desktop/python/matplotlib/newdata.csv')
df.columns
df.dropna(inplace=True)
```

```python

for i in df['LanguageHaveWorkedWith']:
    print(i)
```

```python
count=Counter()
```

```python
for i in df['LanguageHaveWorkedWith']:
    count.update(i.split(';'))
print(count)
languages=[]
popularity=[]
for i in count.most_common(15):
    languages.append(i[0])
    popularity.append(i[1])
print(languages)
print(popularity)
```

```python
plt.pie(popularity,labels=languages,wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
for i in count.most_common(5):
    languages.append(i[0])
    popularity.append(i[1])
print(languages)
print(popularity)
```

#### Explode

```python
# If you want one of the wedges to stand out? The explode parameter allows you to do that
# Each value represents how far from the center each wedge is displayed

languages=['JavaScript','HTML/CSS','SQL','Python','TypeScript']
popularity=[92886,78284,70254,68310,49504]
explode=[0,0,0,0.2,0]
plt.pie(popularity,labels=languages,explode=explode,wedgeprops={'edgecolor':'black'})
plt.show()
```

#### Percentage

```python
# To add percentages to each wedge: use autopct paramenter
plt.pie(popularity,labels=languages,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.show()
```

#### Legend 

```python
plt.pie(popularity,labels=languages,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.tight_layout()
plt.legend(title='languages')
plt.show()
```
