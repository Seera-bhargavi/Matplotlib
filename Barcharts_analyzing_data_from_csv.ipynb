```python
# returning values in a row and making into list using split()

df.loc[3,'LanguageHaveWorkedWith'].split(';')
```

```python
# Counter() function counts the each item from list and returns dictionary.It has built-in method i.e most_common(n). It
# returns a list of top n elements from most common to least common.list consist of tuples and each tuple consist element 
# and its count

count=Counter()
for row in df['LanguageHaveWorkedWith']:
    count.update(row.split(';'))
print(count.most_common(15))
```

#### Plotting Data

```python
# Creating x-values and y-values from the above list i.e
# [('JavaScript', 46443), ('HTML/CSS', 39142), ('SQL', 35127), ('Python', 34155), ('TypeScript', 24752), ('Java', 23644),
# ('Bash/Shell', 20656), ('C#', 19883), ('C++', 16024), ('PHP', 14827), ('C', 13692), ('PowerShell', 8575), ('Go', 7922),
# ('Rust', 6625), ('Kotlin', 6507)]

language = []
lang_count = []
for i in count.most_common(15):
    language.append(i[0])
    lang_count.append(i[1])
print(language)
print(lang_count)
```

#### Creating a Bar chart

```python
# Creating barchart with passing lists
plt.bar(language,lang_count)

# Adding lables
plt.title('Most Popular languages')
plt.xlabel('Programming languages')
plt.ylabel('No of people who use')
```

```python
# when we have large data its better to view in a horizontal manner. to do this use plt.barh()
plt.barh(language,lang_count)
plt.xlabel('No of people who use')
plt.tight_layout()
plt.show()
```

```python
# If you want Most Popular language at top i.e javascript
language.reverse()
lang_count.reverse()
plt.barh(language,lang_count)
plt.xlabel('No of people who use')
plt.tight_layout()
plt.show()
```
